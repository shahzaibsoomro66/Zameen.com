```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("zameen-karachi-dataset.csv")
df.head(5)
```

```python
df.location = np.array([one.replace(', Karachi, Sindh' , '') for one in df.location])
```

```python
df.price.sample(10)
```

```python
temp = []
for one in np.array(df.price):
    sep = one.split(" ")
    if sep[1] == "Lakh":
        temp.append(float(sep[0])*100000)
    elif sep[1] == "Crore":
        temp.append(float(sep[0])*10000000)
    elif sep[1] == "Arab":
        temp.append(float(sep[0])*(100*10000000))

df.price = np.array(temp)
```

```python
df.price.sample(10)
```

```python
df.price = df.price.astype(float).astype(int)
```

```python
toDrop = ["purpose"]
```

```python
df["location"].unique()
```

```python
temp = df['location'].value_counts().index.tolist()
tempLocation = df['location']
for i, one in enumerate(df["location"].value_counts()):
    if int(one) <= 500:
        tempLocation = np.array([jj.replace(temp[i] , 'Others') if jj == temp[i] else jj for jj in tempLocation])
        
# tempLocation.value_counts()
df['location'] = (np.array(tempLocation))
# unique_elements, counts_elements = np.unique(tempLocation, return_counts=True)
# print("Frequency of unique values of the said array:")
# print(np.asarray((unique_elements, counts_elements)))
```

```python
types = pd.get_dummies(df["location"])
df = pd.concat([df,types],axis=1)

df.drop("location",axis=1,inplace=True)

df.head(2)
```

```python
df["type"].unique()
```

```python
types = pd.get_dummies(df["type"])
df = pd.concat([df,types],axis=1)

df.drop("type",axis=1,inplace=True)

df.head(2)
```

```python
df = df.drop(toDrop, axis=1)
```
